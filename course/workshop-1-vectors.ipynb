# Workshop 1 - The World of Vectors

Let’s start with something simple — a **vector**.

So, what even *is* this thing?
$$
v = \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}$$

At first glance, it’s just two numbers stacked on top of each other. But what are they *doing* there?
Can we think of it as **v** having two buddies — ($v_1$) and ($v_2$)?
Or maybe as **v** having two *directions* — one along ($v_1$), another along ($v_2$)?
Or could it even represent something like an image with two pixels?

Turns out, it can be all of these, depending on the universe you’re in. That’s the magic of vectors — they’re simple in form, but their possibilities stretch out to infinity.
